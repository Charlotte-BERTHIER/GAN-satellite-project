```python
def ratio(image_path, threshold=0.90):
    """
    Vérifie si une image a plus de 90% pourcentage de pixels noirs.
    """
    img = Image.open(image_path).convert('L') #On le passe en niveau de gris = un seul canal.
    pixels = list(img.getdata())
    black_pixels = sum(1 for pixel in pixels if pixel == 0)
    total_pixels = len(pixels)
    
    black_ratio = black_pixels / total_pixels
    return black_ratio > threshold
```

```python
def delete_black_images(folder_path, threshold=0.90):
    """
    Supprime les images qui ont plus de 90% pourcentage de pixels noirs.
    """
    for filename in os.listdir(folder_path):
        #print(filename)
        if filename.lower().endswith(('.jpg')):
            file_path= os.path.join(folder_path, filename)
            if ratio(file_path, threshold)==True:
                os.remove(file_path)
                print(f"Supprimé: {file_path}")
```

```python
folder_path = 'data/image'
delete_black_images(folder_path)
```

```python
#dataset_dir = 'data/image'
images = os.listdir()

new_file = '../dataset_double'
if not os.path.exists(new_file):
    os.makedirs(new_file)

for image in images : 
    try : 

        if image.endswith('sat.jpg'):
            print(image[:-8])
            # Charger les deux images
            image1 = cv2.imread(f"{image}")
            image2 = cv2.imread(f"{image[:-7]}occ.jpg")

            # S'assurer que les deux images ont la même hauteur
            min_height = min(image1.shape[0], image2.shape[0])
            image1 = image1[:min_height, :]
            image2 = image2[:min_height, :]

            # Concaténer horizontalement les deux images
            concatenated_image = np.concatenate((image1, image2), axis=1)

            # Enregistrer l'image concaténée
            new_path = os.path.join(new_file, f"{image[:-8]}.jpg")
            cv2.imwrite(new_path, concatenated_image)
    except : 
        print("erreur")
        pass
```
